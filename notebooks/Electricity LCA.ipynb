```python
def get_construction_excs(act):
    plantex = [a for a in act.technosphere() if "plant" in a.input["name"] or
                  "wind turbine" in a.input["name"] or
                  "construction" in a.input["name"] or
                  "unit" in a.input["name"] or
                  "furnace" in a.input["name"]]
    if len(plantex) == 0:
        print("No plant found for {}".format(act))
    else:
        print("Construction activities for {}: {}".format(act, plantex))
        return plantex

def attr_lca(act, *args, **kwargs):
    lca = LCA({act: 1}, *args, **kwargs)
    lca.lci()
    lca.lcia()
    constr_score = 0
    for ex in get_construction_excs(act):
        clca = LCA({ex.input: ex.amount}, *args, **kwargs)
        clca.lci()
        clca.lcia()
        constr_score += clca.score
    return {
        "construction": constr_score,
        "operation": lca.score - constr_score}
```

```python
attr_lca(coal, method=met)
```
